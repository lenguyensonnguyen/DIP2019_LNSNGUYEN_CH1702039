Import các thư viện cần thiết

```python
import numpy as np
from skimage import color, data, restoration

import cv2
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.rcParams['figure.dpi'] = 500
```

đọc và show ảnh

```python
img = cv2.imread('./Anh nhieu/5n.jpg', 0)
```

```python
plt.imshow(img, cmap='gray')
```

vẽ histogram

```python
plt.hist(img.ravel(), 256, [0, 256])
plt.show()
```

THỬ CÂN BẰNG HISTOGRAM VÀ SHOW KẾT QUẢ

ĐỌC VÀ HIỂN THỊ ẢNH GỐC

```python
anh_goc = cv2.imread('./Anh nhieu/5g.jpg', 0)
plt.imshow(anh_goc, cmap='gray')
```

VẼ HISTOGRAM ẢNH GỐC

```python
plt.hist(anh_goc.ravel(), 256, [0, 256])
plt.show()
```

DÙNG KỸ THUẬT MATCH_HISTOGRAM ĐỂ CHO ẢNH KẾT QUẢ TỐT HƠN

```python
from skimage.exposure import match_histograms
```

```python
img_match = match_histograms(img, anh_goc, multichannel=False)
plt.imshow(img_match, cmap='gray')
```

HISTOGRAM SAU KHI MATCHING

```python
plt.hist(img_match.ravel(), 256, [0, 256])
plt.show()
```

```python
plt.imsave('./Anh nhieu/5_denoise.jpg', img_match, cmap='gray')
```

TÍNH CÁC MSE

```python
import math
from skimage import img_as_float
from skimage.measure import compare_mse as mse
```

```python
mse_noise = math.sqrt(mse(img_as_float(img), img_as_float(anh_goc)))
mse_noise
```

```python
mse_match = math.sqrt(mse(img_as_float(img_match), img_as_float(anh_goc)))
mse_match
```
